# Double Pendulum


## Introduction
Let us approach a particularly interesting problem, the double pendulum. The double pendulum system we will explore exhibits chaotic behavior.

## Prerequisites
Part of the battle, you will find, is in deriving the differential equations to model the system in question. Therefore, it is useful to make a few notes on notation and methods.

### Notation
It is common to denote the time derivative of a time-dependent quantity with a dot over it. For example, $\dot q$ is shorthand for $\frac{dq}{dt}$. This is read as "q dot". Similarly, $\frac{d^{2}q}{dt^{2}}$ is expressed as $\ddot q$, and this is read "q double dot".


  

### Lagrange's System of Mechanics
Sometimes, it is more useful to solve a problem by finding a trajectory. More specifically, given a set of initial conditions, we find the path of least action. This is fulfilled by Lagrangian Mechanics, with the flagship formula to use being the Euler-Lagrange equation: 
$\frac{d}{dt}(\frac{\partial L}{\partial \dot q_i}) - \frac{\partial L}{\partial q_i} = 0$ for all $i$ degrees of freedom, where $L$ is defined as $L = T - V$, with $T$ being the Kinetic Energy of the system, and $V$ being the Potential Energy. In addition, a generalized (not necessarily Cartesian) system of coordinates may be utilized. For a system to be solved by the Euler-Lagrange Equation, it must meet the following criteria: 
   
1. Independence - the system is able to move when all but one of its coordinates are fixed  
2. Completeness - chosen coordinates exist without disappearing in the system
3. Holonomicity - the number of degrees of freedom in the system matches the number of coordinates  

This description is intentionally vague, simply because the focus is not on Lagrangian Mechanics. It is highly recommended to formally look into Lagrangian Mechanics, and pay close attention to the derivation of the Euler-Lagrange Equation.     

## Derivation of Differential equations  
### The Problem
A massless rod of length $l_1$ is attached to the wall, and is free to rotate around its fixed point. At its end is a point mass $m_1$. So far, this is identical to the setup for a simple pendulum. Attached to $m_1$ is another massless rod of length $l_2$, with a mass $m_2$ attached to its end. This second rod is free to rotate about $m_2$. One may find it useful to look at an image of a setup of the double pendulum.
  
### The Approach
We start by setting up a Cartesian System of coordinates for this problem. We choose the origin to be the fixed point of $l_1$. Now, we choose to describe the position of $m_1$ with Cartesian coordinates $x_1$ and $y_1$. Defining $X_2$ and $Y_2$ as the position of $m_2$ with respect to $m_1$, we can write the position of $m_2$ with respect to the origin as $x_2 = x_1 + X_2$ and $y_2 = y_1 + Y_2$ (do note the difference between $x_2$ and $X_2$).  
We intend to approach this situation using the Euler-Lagrange Equation. Therefore, we test the conditions. 

#### Independence
When checking for independence, it helps to imagine a single pendulum (with who degrees of freedom, $x_1$ and $y_1$). If we were to fix $x_1$, the system would be unable to move. In other words, $y_1$ cannot change without a change in $x_1$. This is the same reason why a circle in rectangular coordinates is dependent on both its $x$ and $y$ components (consider how this relates to the trajectory of the single pendulum). The same argument invalidating the compatability of the Cartesian single pendulum with the Euler_Lagrange equation can be translated to the double pendulum. So the bottom line is that, before going forward, we must choose a different set of coordinates that will allow us to meet the condition of independence. We first test polar coordinates. Instead of $x_1$ and $y_1$ for $m_1$, and $x_2$ and $y_2$ for $m_2$, we can use $\theta_1$ and $\theta_2$ for $m_1$ and $m_2$, respectively. We measure $\theta_1$ from the fixed point of the rod connecting $m_1$ to its pivot, and $\theta_2$ about $m_1$ (the fixed point of the rod connecting to $m_2$). Now we test for independence again. Fixing $\theta_1$ still allows $\theta_2$ to change, and vice versa. As a result, the first condition for Lagrangian Mechanics is satisfied. 

#### Completeness
Next, we check for completeness. Based on the definition, this is pretty straightforward. There is nothing in the system that makes the chosen coordinates disappear at certain points in time, so the second condition is satisfied.

#### Holonomicity
The system has two degrees of freedom about each pivot point, and two coordinates to describe them. Therefore, the third and final condition is satisfied, and we are free to approach the problem using Lagrangian Mechanics.

# Solution  
For the Euler-Lagrange, we need to first find equations for the kinetic and potential energies. Kinetic energy for this system, denoted $T$, takes the form $\frac{1}{2} m_1 \dot q_1 + \frac{1}{2} m_2 \dot q_2$, where $\dot q_i$ is the velocity of mass $$m_i$.  However, we can also express this as the sum of the velocity of each coordinate, bringing the equation to $T = \frac{1}{2} m_1 (\dot x_1 ^ {2} + \dot y_1 ^ {2})  + \frac{1}{2} m_2 (\dot x_2 ^ {2} + \dot y_2 ^ {2})$. Similarly, potential energy, denoted $V$, is $V = m_1 g y_1 + m_2 g y_2$. Before we can do anything with this, we have to convert to polar.

We measure $\theta_1$ from the vertical. We can now write the $x_1$ as $l_1\sin\theta_1$, and $y_1$ as $-l_1\cos\theta_1$ (given our corrdinate system). We measure $\theta_2$ from the vertical as well. Reviewing how we chose to define the position of $m_2$, it should be very clear why the polar equivalent is $l_1\sin\theta_1 + l_2\sin\theta_1$ for $x_2$, and $-l_1 \cos\theta_1 - l_2\cos\theta_2$ for $y_2$.  

In order to write the Kinetic Energy, let's first differentiate each of our coordinates with respect to time.  
Starting with $x_1$, $\frac{d}{dt} x_1 = l_1 \frac{d}{dt} \sin \theta_1$.  
Using the chain rule, this is simply $l_1 \cos (\theta_1) \dot \theta_1$.  
Utilize a similar approach for $\dot y_1$. The result should be $l_1 \sin(\theta_1) \dot \theta_1$.  
  
Now, $x_2$: $\frac{d}{dt} x_2 = \frac{d}{dt} x_1 + \frac{d}{dt} l_2 \sin \theta_2 = l_1 \cos (\theta_1) \dot \theta_1 + l_2 \cos (\theta_2) \dot \theta_2$.  

The same approach for $$y_2$ returns $\dot y_2$ should yield $l_1 \sin (\theta_1) \dot \theta_1 + l_2 \sin (\theta_2) \dot \theta_2$.

## Kinetic Energy
We can write the KE for the system as $T = T_1 + T_2$, where $T_1$ is the KE of the first rod, and $T_2$ is the KE of the second rod.  


Let's start with $T_1$:  
$T_1 = \frac{1}{2}m_1 (\dot{x_1}^2 \dot{y_1}^2) = \frac{1}{2}m_1(l_1^{2}\cos^{2}\theta_1 \dot{\theta_1^{2}} + l_1^{2}\sin^{2}\theta_1 \dot{\theta_1^{2}})$  
Simplifying a little,  
$\frac{1}{2}m_1 l_1^{2}\dot{\theta_1^{2}}(\cos^{2} + \sin^{2})$.  
With a healthy dose of trigonometric identities, namely, $(\cos^{2} \alpha + \sin^{2} \alpha) = 1$,  
$T_1 = \frac{1}{2}m_1 l_1^{2}\theta_1^{2}$.  
  
We have a little more work for $T_2$:  
$T_2 = \frac{1}{2}m_2((l_1\cos\theta_1 \dot{\theta_1} + l_2\cos\theta_2 \dot{\theta_2})^{2} + (l_1^{2}\sin^{2}\theta_1 \dot{\theta_1} + l_2^{2}\sin^{2}\theta_2 \dot{\theta_2})^{2})$.  
Expanding,  
$\frac{1}{2}m_2[(l_1^{2}\cos^{2}\theta_1 (\dot{\theta_1}^{2}) + l_2^{2}\cos^{2}\theta_2 (\dot{\theta_2}^{2}) + 2 l_1 l_2 (\cos \theta_1)(\cos \theta_2) \dot{\theta_1}\dot{\theta_2}) + (l_1^{2}\sin^{2}\theta_1 (\dot{\theta_1}^{2}) + l_2^{2}\sin^{2}\theta_2 (\dot{\theta_2}^{2}) + 2 l_1 l_2 (\sin \theta_1)(\sin \theta_2) \dot{\theta_1}\dot{\theta_2})]$.  
Taking a look at the equation, there's a little of work we can do to simplify:  
$\frac{1}{2} m_2 [l_1^{2} \dot{\theta_1}^{2} (\cos^{2} \theta_1 + \sin^{2} \theta_1) + l_2^{2} \dot{\theta_2}^{2} (\cos^{2} \theta_2 + \sin^{2} \theta_2) + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}((\cos \theta_1)(\cos \theta_2) + (\sin \theta_1)(\sin \theta_2))]$.  
Now, let's reap a benefit of paying attention in algebra class. Using two trignometric identites,  $(\cos^{2} \alpha + \sin^{2} \alpha) = 1$, and $\cos\alpha\cos\beta + \sin\alpha\sin\beta = cos(\alpha - \beta)$ we can make this a lot prettier.  
$\frac{1}{2} m_2 [l_2^{2} \dot{\theta_1}^{2} (1) + l_2^{2} \dot{\theta_2}^{2} (1) + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}(\cos(\theta_1 - \theta_2))]$. 
So 
$T_2 = \frac{1}{2} m_2 [l_2^{2} \dot{\theta_1}^{2} + l_2^{2} \dot{\theta_2}^{2} + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}(\cos(\theta_1 - \theta_2))]$.  
  
Therefore, for $T$, we have
$T_1 + T_2 = \frac{1}{2}m_1 l_1^{2}\dot{\theta_1^{2}} + \frac{1}{2} m_2 [l_2^{2} \dot{\theta_1}^{2} + l_2^{2} \dot{\theta_2}^{2} + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}(\cos(\theta_1 - \theta_2))]$.  
  
## Potential Energy  
We derive the potential energy, $V$, with 3 things in mind:  
1. A visual image of the system itself  
2. The coordinate system  
3. The fact that Gravitational Potential Energy can be calculated through the formula $V = mgy$.  
  
Again, we can consider $V = V_1 + V_2$.  
  
For $V_1$, we note that the mass in question is $m_1$. It is acted on by the gravitational constant, $g$, and the $y$ direction in polar coordinates is $-l_1 \cos \theta_1$.   
For $V_2$, the force is $m_2g$, and acting along the $y$ direction, which is $-(l_1\cos\theta_1 + l_2 \cos \theta_2)$.  
  
$V = -l_1 \cos \theta_1 -(l_1\cos\theta_1 + l_2 \cos \theta_2)$, or
$-(m_1 + m_2)gl_1 \cos \theta_1 - m_2 g l_2 \cos \theta_2$.  

## The Lagrangian  
Keep in mind, $L = T - V$.  
Substituting, we write the Lagrangian as  
$L = [\frac{1}{2}m_1 l_1^{2}\dot{\theta_1^{2}} + \frac{1}{2} m_2 [l_2^{2} \dot{\theta_1}^{2} + l_2^{2} \dot{\theta_2}^{2} + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}(\cos(\theta_1 - \theta_2))] - [-(m_1 + m_2)gl_1 \cos \theta_1 - m_2 g l_2 \cos \theta_2]]$.  
  
Note that while we do formally write out the Lagrangian, it is very helpful to keep $T$ and $V$ in mind.

## Applying the Euler-Lagrange Equation  
We apply the Euler-Lagrange Equation, $\frac{d}{dt} (\frac{\partial L}{\partial \dot q_1}) - \frac{\partial L}{\partial q_1} = 0$, for $q_1 = \theta_1, \theta_2$.


### For $\theta_1$  
$\frac{d}{dt} (\frac{\partial L}{\partial \dot \theta_1}) - \frac{\partial L}{\partial \theta_1} = 0$.  
Notice that $\frac{\partial L}{\partial \dot{\theta_1}} = \frac{\partial T}{\partial \dot{\theta_1}}$, because $V$ does not change with respect to $\dot{\theta_1}$.  
First, we take the partial derivative of the expression with respect to $\dot{\theta_1}$:  

$\frac{\partial L}{\partial \dot{\theta_1}} = \frac{\partial T}{\partial \dot{\theta_1}}$

$= \frac{\partial}{\partial \dot{\theta_1}}[\frac{1}{2}m_1 l_1^{2} \dot{\theta_1}^{2} + \frac{1}{2}m_2[l_1^{2} \dot{\theta_1}^{2} + l_1^{2} \dot{\theta_2}^{2} + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2} (\cos(\theta_1 - \theta_2))]]$  
  
Taking the derivative with respect to $\dot{\theta_1}$:  
  
$ = m_1 l_1^{2} \dot{\theta_1} + \frac{1}{2}m_2[2l_1^{2} \dot{\theta_1} + 0 + 2 l_1 l_2 \dot{\theta_2} (\cos(\theta_1 - \theta_2))]$  
  
Now, a similar process for $t$: 
$\frac{d}{dt}(m_1 l_1^{2} \dot{\theta_1} + \frac{1}{2}m_2[2l_1^{2} \dot{\theta_1} + 0 + 2 l_1 l_2 \dot{\theta_2} (\cos(\theta_1 - \theta_2))]) =$
  
$m_1 l_1^{2} \ddot{\theta_1} + \frac{1}{2}m_2[2l_1^{2} \ddot{\theta_1}+ 2 l_1 l_2(-\dot{\theta_2} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2}) + \cos(\theta_1 - \theta_2)\ddot{\theta_2}] = $  
  
$m_1 l_1^{2} \ddot{\theta_1} + m_2l_1^{2} \ddot{\theta_1}- m_2 l_1 l_2 \dot{\theta_2} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2}) + m_2 l_1 l_2 \cos(\theta_1 - \theta_2)\ddot{\theta_2}$  

Ensure to use the chain rule for the above!  

Now, let us calculate $-\frac{\partial L}{\partial \theta_1}$:  

$-\frac{\partial}{\partial \theta_1} [\frac{1}{2}m_1 l_1^{2}\dot{\theta_1^{2}} + \frac{1}{2} m_2 [l_2^{2} \dot{\theta_1}^{2} + l_2^{2} \dot{\theta_2}^{2} + 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}(\cos(\theta_1 - \theta_2))] - [-(m_1 + m_2)gl_1 \cos \theta_1 - m_2 g l_2 \cos \theta_2]]$  

$= -\frac{\partial}{\partial \theta_1}[m_2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}(\cos(\theta_1 - \theta_2))+ (m_1 + m_2)gl_1 \cos \theta_1 + m_2 g l_2 \cos \theta_2]$  

$= -[-m_2 l_1 l_2 \dot{\theta_1} \dot{\theta_2}(\sin(\theta_1 - \theta_2))- (m_1 + m_2)gl_1 \sin \theta_1]$  

$= m_2 l_1 l_2 \dot{\theta_1} \dot{\theta_2} \sin(\theta_1 - \theta_2) + (m_1 + m_2)gl_1 \sin \theta_1$  

Now we are free to add our two results and set the sum equivalent to 0 (Euler-Lagrange):

$m_1 l_1^{2} \ddot{\theta_1} + m_2l_1^{2} \ddot{\theta_1}- m_2 l_1 l_2 \dot{\theta_2} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2}) + m_2 l_1 l_2 \cos(\theta_1 - \theta_2)\ddot{\theta_2} + m_2 l_1 l_2 \dot{\theta_1} \dot{\theta_2} \sin(\theta_1 - \theta_2) + (m_1 + m_2)gl_1 \sin \theta_1 = 0$  
  
$= m_1 l_1 \ddot{\theta_1} + m_2 l_1 \ddot{\theta_1}- m_2 l_2 \dot{\theta_2} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2}) + m_2 l_2 \cos(\theta_1 - \theta_2)\ddot{\theta_2} + m_2 l_2 \dot{\theta_1} \dot{\theta_2} \sin(\theta_1 - \theta_2) + (m_1 + m_2)g \sin \theta_1$
  
$= (m_1 + m_2)l_1 \ddot{\theta_1} + (m_1 + m_2)g\sin \theta_1+ m_2 l_2 \ddot{\theta_2}\cos(\theta_1 - \theta_2)+ m_2l_2 \dot{\theta_2}^{2}\sin(\theta_1 - \theta_2) = 0$



### For $\theta_2$  

For this we once again apply the Euler-Lagrange equation, modifying our approach slightly to account for $\theta_2$.  

$\frac{d}{dt} \frac{\partial L}{\partial \dot{\theta_2}} - \frac{\partial L}{\partial \theta_2} = 0$
  
$\frac{\partial L}{\partial \dot{\theta_2}} = \frac{\partial}{\partial \dot{\theta_2}} [\frac{1}{2} m_1 l_1 \dot{\theta_1}^{2} + \frac{1}{2}m_2[l_1^{2} \dot{\theta_1}^{2} + l_2^{2} \dot{\theta_2}^{2}+ 2 l_1 l_2 \dot{\theta_1} \dot{\theta_2} \cos(\theta_1 - \theta_2)]]$
  
$= 0 + \frac{1}{2} m_2[0 + 2 l_2^{2} \dot{\theta_2} + 2 l_1 l_2 \dot{\theta_1} \cos(\theta_1 - \theta_2)]$

Now,  
$\frac{d}{dt}[\frac{1}{2} m_2[2 l_2^{2} \dot{\theta_2} + 2l_1 l_2 \dot{\theta_1} \cos(\theta_1 - \theta_2)]]$

$= \frac{1}{2} m_2 [2 l_2^{2} \ddot{\theta_2} + 2 l_1 l_2 (-\dot{\theta_1} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2})+ \cos(\theta_1 - \theta_2)(\ddot{\theta_1})]$

$= m_2 l_2^{2} \ddot{\theta_2}- m_2 l_1 l_2 \dot{\theta_1} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2})+ m_2 l_1 l_2 \ddot{\theta_1} \cos(\theta_1 - \theta_2)$
  
The other part:

$\frac{\partial L}{\partial \theta_2} = \frac{\partial}{\partial \theta_2}[m_2 l_1 l_2 \dot{\theta_1} \dot{\theta_2} \cos(\theta_1 - \theta_2) +(m_1 + m_2) g l_1 \cos(\theta_1) + m_2 g l_2 \cos(\theta_2)$

$= m_2 l_1 l_2 \dot{\theta_1}\dot{\theta_2} \sin(\theta_1 - \theta_2) - 0 - m_2 g l_2 \sin \theta_2$

Finally,  

$m_2 l_2^{2} \ddot{\theta_2} - m_2 l_1 l_2 \dot{\theta_1} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2}) + m_2 l_1 l_2 \ddot{\theta_1} \cos(\theta_1 - \theta_2) - m_2 l_1 l_2 \dot{\theta_1}\dot{\theta_2} \sin(\theta_1 - \theta_2) + m_2 g l_2 \sin \theta_2 = 0$

Canceling out common terms:  
$ l_2 \ddot{\theta_2} - l_1 \dot{\theta_1} \sin(\theta_1 - \theta_2)(\dot{\theta_1} - \dot{\theta_2}) + l_1 \ddot{\theta_1} \cos(\theta_1 - \theta_2) - l_1 \dot{\theta_1}\dot{\theta_2} \sin(\theta_1 - \theta_2) + g \sin \theta_2 = 0$

We end up with:  

$l_2 \ddot{\theta_2} + g \sin(\theta_2) + l_1 \ddot{\theta_1} \cos(\theta_1 - \theta_2) - l_1 \dot{\theta_1}^{2} \sin(\theta_1 - \theta_2) = 0$














```python
import numpy as np 
import matplotlib.pyplot as plt 


#CONSTANTS
g = 9.8 #m/s^2

m1 = 10 #kg
m2 = 10 #kg

l1 = 2 #m
l2 = 1 #m

#INITIAL CONDITIONS
THETA_1_0 = np.pi / 4
THETA_2_0 = np.pi / 4

THETA_DOT_1_0 = 0
THETA_DOT_2_0 = 0

THETA_DOUBLE_DOT_1_0 = 0
THETA_DOUBLE_DOT_2_0 = 0



#POLAR TO RECTANGULAR
def get_x1(theta_1):
    return l1 * np.sin(theta_1)

def get_y1(theta_1):
    return -l1 * np.cos(theta_1)

def get_x2(theta_1, theta_2):
    return get_x1(theta_1) + l2 * np.sin(theta_2)

def get_y2(theta_1, theta_2):
    return get_y1(theta_1) - l2 * np.cos(theta_2)

#RANGE REDUCTION
RANGE = 2 * np.pi
def toRange(angle):
    angle = angle % RANGE
    if angle > np.pi:
        angle -= RANGE
    return angle
    


#ODEs
def get_theta_double_dot_1(theta_1, theta_2, theta_double_dot_2, theta_dot_2):
    eq = -((m1 + m2)*g*np.sin(theta_1) + m2*l2*theta_double_dot_2*np.cos(theta_1-theta_2) + m2*l2*(theta_dot_2**2)*np.sin(theta_1 - theta_2)) /  (m1 + m2)
    return eq

def get_theta_double_dot_2(theta_2, theta_1, theta_dot_1, theta_double_dot_1):
    eq = (-(g*np.sin(theta_2) + (l1*theta_double_dot_1*np.cos(theta_1-theta_2))) + (l1 * theta_dot_1*np.sin(theta_1-theta_2)))/l1
    return eq
    
    
#ANGLES
def theta(t_0, t_f):
    theta_1 = THETA_1_0
    theta_2 = THETA_2_0

    theta_dot_1 = THETA_DOT_1_0
    theta_dot_2 = THETA_DOT_2_0

    theta_double_dot_1 = THETA_DOUBLE_DOT_1_0
    theta_double_dot_2 = THETA_DOUBLE_DOT_2_0

   

    dt = 0.001
    timeData = np.arange(t_0, t_f, dt)

    theta_1_vals = []
    theta_2_vals = []

    theta_dot_1_vals = []
    theta_dot_2_vals = []

    theta_double_dot_1_vals = []
    theta_double_dot_2_vals = []

    x1_vals = []
    y1_vals = []

    x2_vals = []
    y2_vals = []

    for time in timeData:
        theta_double_dot_1 = get_theta_double_dot_1(theta_1, theta_2,theta_double_dot_2, theta_dot_2)

        theta_double_dot_1_vals.append(theta_double_dot_1) 
        

        theta_double_dot_2 = get_theta_double_dot_2(theta_2, theta_1, theta_dot_1, theta_double_dot_1)
    
        theta_double_dot_2_vals.append(theta_double_dot_2) 


        theta_dot_1 += theta_double_dot_1 * dt
        theta_dot_1_vals.append(theta_dot_1)

        theta_dot_2 += theta_double_dot_2 * dt 
        theta_dot_2_vals.append(theta_dot_2)

        theta_1 += (theta_dot_1 * dt % (2 * np.pi))
        #theta_1 = toRange(theta_1)
        theta_1_vals.append(theta_1)
        
        
        theta_2 += (theta_dot_2 * dt) 
        #theta_2 = toRange(theta_2)
        theta_2_vals.append(theta_2)

        x1_vals.append(get_x1(theta_1))
        y1_vals.append(get_y1(theta_1))

        x2_vals.append(get_x2(theta_1, theta_2))
        y2_vals.append(get_y2(theta_1, theta_2))
    
    for i in range(0, len(theta_1_vals)):
        theta_1_vals[i]= toRange(theta_1_vals[i])
        theta_2_vals[i]= toRange(theta_2_vals[i])
        


    
    #print(theta_double_dot_1_vals)
    #print(theta_dot_1_vals)
    #print(theta_1_vals)
    

    plt.figure(1)

    plt.plot(timeData, theta_1_vals, label="theta_1 vs. time")
    plt.figure(2)
    plt.plot(timeData, theta_2_vals, label="theta_2 vs. time")
    plt.figure(3)
    plt.plot(theta_1_vals, theta_dot_1_vals, label="theta_2 vs theta_1")
    plt.figure(4)
    plt.plot(x2_vals, y2_vals)
    

    #return theta_1, theta_2


    



theta(0, 22)

    






```
